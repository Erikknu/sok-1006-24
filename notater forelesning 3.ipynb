### 1. Elastisitetsbegrepet

Elastisitet gir oss et mål på hvor følsomt tilbud og etterspørsel er for prisendringer. Vi har tidligere tegnet disse funksjonene som en rett linje, og helningen til linjen forteller oss noe om hvordan mengde tilbudt/etterspurt reagerer på prisendringer. Tenk at vi har invers etterspørsel etter laks i Norge gitt ved $p = 300 - 0.25x$. Her er prisen målt i NOK og mengde i kilogram. Vi ser at for å selge en ekstra kilo med laks i Norge må prisen reduseres med 25 øre. I Spania er etterspørsel etter laks gitt ved $q = 500 - 0.3x$, og q er nå prisen i EUR. For å selge en ekstra kilo laks til det spanske markedet må prisen reduseres med 30 cent. For å sammenlikne mellom Norge og Spania må vi ha informasjon om valutakursen mellom NOK og EUR. Dette eksempelet illustrerer at det å bruke helningen til etterspørsel/tilbud for å måle prisfølsomhet er avhengig av måleenheten for verdi. Sammenlikningen vil også påvirkes av hvordan vil måler kvantum (kilo, ton osv).     

Elastisitet er målt i prosent og er da uavhengig av måleenheten for verdi og mengde. Etterspørselselastisiteten måler hvor mange prosent etterspørselen endres når prisen øker med 1%. Tenk at vi har etterspørsel $X^D(p)$. En liten endring i etterspørsel kan vi kalle $dX^D$, slik at en prosentvis endring i etterspørsel er
$$prosentvis\; endring \;i \; etterspørsel = \frac{dX^D}{X^D}\tag{1}.$$

La $dp$ vise en liten endring i pris, og vi kan skrive den prosentvise endringen i pris som 
$$prosentvis\; endring \;i \; pris = \frac{dp}{p}\tag{2}.$$

Etterspørselens priselastisitet er gitt ved $\epsilon^D$ som er den prosentvise endringen i kvantum etterspurt delt på den prosentvise endringen i pris.
$$\epsilon^D = \frac{prosentvis\; endring \;i \; etterspørsel}{prosentvis\; endring \;i \; pris}.$$
Tenk på et talleksempel før vi utleder dette formelt. Vi undersøker etterspørsel i et marked og har funnet at etterspørsel har nylig falt med 6%; i samme periode har prisen økt med 3%. Deler vi den ene på den andre så finner vi at -6%/3% = -2% (minus ettersom etterspørselen har falt). Tolkningen nå er når prisen øker med 1% så faller etterspørsel med 2%. Det er nettopp dette som etterspørselens priselastisitet måler!

Ved å bruke likning (1) og (2) kan vi skrive
$$\epsilon^D = \frac{\frac{dX^D}{X^D}}{\frac{dp}{p}}.$$
Her deles en brøk på en annen brøk som vi kan forenkle som følger
$$\epsilon^D = \frac{\frac{dX^D}{X^D}}{\frac{dp}{p}}=\frac{dX^D}{X^D}\frac{p}{dp}.$$

Som vi kan skrive om som:
$$\epsilon^D = \frac{\frac{dX^D}{X^D}}{\frac{dp}{p}}=\frac{dX^D}{X^D}\frac{p}{dp}=\frac{dX^D}{dp}\frac{p}{X^D}.$$

Legg merke til at $\frac{dX^D}{dp}$ er helningen til den direkte etterspørselsfunksjonen $X^D$. Ettersom etterspørselen heller nedover er $\frac{dX^D}{dp} < 0$ og dermed er $\epsilon^D < 0.$

### Lineær etterspørsel

La oss tenke oss følgende lineære etterspørselsfunksjon:    

$X^D = 500- 5p$ i punkt $p=50, X^D=250.$

La oss tegne dette først.


```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def demand(p):
    return 500-5*p

p_num=np.linspace(0.1,99.9,100)

#lag en akse

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris (p)', loc='top')
                ax.set_xlabel('Kvantum (x)', loc='right')
                ax.set(xlim=(0,400))
                ax.set(ylim=(0,150))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

# plott funksjonen

ax.plot(demand(p_num),p_num, label='Etterspørsel')


# tittel
ax.set_title('Lineær etterspørsel')

#vis navnene:
ax.legend(loc='best',frameon=False);
```

Legg merke til at verdien av elastisiteten i vårt eksempel er avhengig av p og x, dvs vi får forskjellige målinger for prisfølsomhet avhengig av hvor vi er på etterspørselskurven. I eksempelet har vi
$$\epsilon^D = -5\frac{p}{x}.$$

Når $\frac{p}{x}$ er et høyt tall (dvs vi er langt oppe til venstre på etterspørselskurven) får vi et høyt negativt tall for elastisiteten; når $\frac{p}{x}$ er et lavt tall (dvs vi er langt nede til høyre på etterspørselskurven) får vi et relativt lavt negativt tall for elastisiteten.

```python
# definer elastisitet
#legg merke til at dette er definert kun som funksjon av p
def elast_d(p):
    return -5*p/(500-5*p)

# lag en akse med 2 paneler

def create_ax2():
                fig, axs = plt.subplots(1,2, sharey=True, figsize=(15,5))
                for i in range(2):
                    axs[i].set_ylabel('Pris (p)', loc='top')
                    axs[i].spines['top'].set_color('none')
                    axs[i].spines['right'].set_color('none')
                return fig, axs

fig, axs = create_ax2()


axs[0].set(xlim=(0,500))
axs[1].set(ylim=(0,100))
axs[1].set(xlim=(-5,0))
axs[0].set_xlabel('Kvantum (x)', loc='right')
axs[1].set_xlabel('Etterspørselselastisitet', loc='right')

# plott funksjonenen

axs[0].plot(demand(p_num),p_num, label= 'etterspørsel')
axs[1].plot(elast_d(p_num),p_num, label='Etterspørselselastisitet')

# legg inn noen linjer

axs[0].plot ([0, 250], [50, 50], linestyle='dashed', color='black')
axs[0].plot ([250, 250], [0, 50], linestyle='dashed', color='black')

axs[1].plot ([-5, -1], [50, 50], linestyle='dashed', color='black')
axs[1].plot ([-1, -1], [0, 50], linestyle='dashed', color='black')

# vis forskellige verdier i første panel

axs[0].annotate(r'$\epsilon^D < -1$', xy=(120,80), fontsize=15)
axs[0].annotate(r'$\epsilon^D > -1$', xy=(330,40), fontsize=15)
axs[0].annotate(r'$\epsilon^D=-1$', xy=(250, 50), xytext=(260, 60), fontsize=15, arrowprops={"arrowstyle":"->", "color":"gray"})


#vis navnene:
axs[0].legend(loc='best',frameon=False)
axs[1].legend(loc='best',frameon=False)


# tittel 
axs[0].set_title('Etterspørsel')
axs[1].set_title('Etterspørselselastisitet')
fig.suptitle('Lineær etterspørsel');


```

### Lineært tilbud

Vi kan gjøre det samme for tilbudsfunksjonen $X^S(p)$. Tilbudselastisiteten viser hvor mange prosent tilbudet øker når prisen går opp med 1%, og skrives som
$$\epsilon^S=\frac{dX^S}{dp}\frac{p}{x^S}.$$
Ettersom tilbudet heller oppover har vi $\frac{dX^S}{dp}>0$ og dermed $\epsilon^S>0.$     
Ta et eksempel med $X^S=-20+1.5p$. Vi kan regne ut at elastisiteten til tibudet er 
$$\epsilon^S=1.5 \frac{p}{X^S}=1.5 \frac{p}{-20+1.5p}.$$

```python
def tilbud(p):
    return -20 + 1.5*p

def elast_t(p):
    return 1.5*p/(-20+1.5*p)

p_ny=np.linspace(14,100,100)

# bruk fellesaksen vi laget i ste
fig, axs = create_ax2()


axs[0].set(xlim=(0,200))
axs[1].set(ylim=(0,100))
axs[1].set(xlim=(0,3))
axs[0].set_xlabel('Kvantum (x)', loc='right')
axs[1].set_xlabel('Tilbudselastisitet', loc='right')

# plott funksjonenen

axs[0].plot(tilbud(p_ny),p_ny, label= 'Tilbud')
axs[1].plot(elast_t(p_ny),p_ny, label='Tilbudselastisitet')


# vis forskellige verdier i første panel


axs[0].annotate(r'$\epsilon^S faller$', xy=(130, 90), xytext=(55, 45), fontsize=15, arrowprops={"arrowstyle":"->", "color":"gray"})


#vis navnene:
axs[0].legend(loc='best',frameon=False)
axs[1].legend(loc='best',frameon=False)


# tittel 
axs[0].set_title('Tilbud')
axs[1].set_title('Tilbudselastisitet')
fig.suptitle('Lineært tilbud');
```
