#### Sammendrag

Boliger i tromsø har økt i pris og antall salg har gått ned, denne rapporten vil forklare de grunnleggende økonomiske mekanismene som forårsaker dette. Den vil dermed vise hvordan boligmarkedet har utviklet seg og fremgangsmetoden brukt for å lage modellen. Gå over nøkkeltall og til slutt komme med og drøfte forslag til eventuelle tiltak som kan bli satt i bruk. 

```python

import numpy as np
from matplotlib import pyplot as plt
from matplotlib import dates as dt
import sympy as sp 
import pandas as pd
import seaborn as sns
from matplotlib.ticker import FuncFormatter
```

```python
def demand(x):
    return 10-(0.1)*x

def supply(x):
    return 8+(0.15)*x
def new_supply(x):
    return 7.5+(0.15)*x
    
p_num=np.linspace(0,20,20)
```

### Hvorfor boligmarkedet er viktig og hva det innebærer
Boligmarkedet er viktig både for arbeidsmarked og befolkningen. For befolkningen er det viktig å ha boliger som gir ly for vær og vinn, og kan gi gode forhold for glede. For arbeidsmarkedet er det viktig å ha mobile arbeidere som kan flytte dit det er tilgjengelig arbeid. Dette kan ikke skje vis det ikke finnes tilgjengelige boliger hvor det trengs som møter de nødvendige kravene til arbeideren. I tillegg må disse boligene være billige nok til at arbeiderene har rå til å leie eller kjøpe dem. 

### Metode
Mediehuset ITromsø samlet data over boligmarkedet fra 2019 til slutten av 2023. Dataen inkluderer antall salg, salgspris, boligens størrelse og lokasjon og datoen til salget. Dette datasettet vil være basisen på min analyse. En mulig feil i analysen kan komme fra at datasettet ikke inkluderer desember i 2023. Figurene som er illustrert kan dermed være unøyaktige når det innebærer år 2023, dette er tatt i betraktning. Jeg vil basere analysen på salg av bolig og ser bort i fra leie.  

### Modell av boligmarkedet

Boligmarkedet som alle marked er drevet av tilbud og etterspørsel. Etterspørsel i boligmarkedet er enkelt sagt, hvor mange personer som ønsker å kjøpe en bolig i Tromsø. Mens tilbud vil være antall boliger tilgjengelig. Disse faktorene sammensatt gir mekanismen som bestemmer prisen. F.eks hvis det finnes 8 boliger men 10 personer i søk om bolig vil de konkuere om pris sånn at de 8 som er villig til å betale mest kjøper boligene. Dette kan visualiseres med figuren under. 

```python

plt.plot(demand(p_num),color = "green")
plt.axvline(x = 7, color = "orange")

plt.xlim(0,20)


plt.plot ([0,7], [9.25, 9.25], linestyle='dashed', color='black')

plt.annotate("Tilbud",xy=(5, 8.5), annotation_clip=False,color = "orange")
plt.annotate("Etterspørsel",xy=(3, 9.7), annotation_clip=False,color = "green")
plt.annotate("Pris",xy=(.2, 9.3), annotation_clip=False)

plt.ylabel('Pris (p)', loc ='top', rotation = 0)
plt.xlabel('Kvantum (x)', loc = 'right')

plt.title('Markedslikevekt figur ved statisk tilbud')
```

Når tilbudet øker går den Orange tilbuds linjen ned på den grønne etterspørselslinjen, dermed går prisen ned






Tilbudskurven på kort tid kan vises vertikal siden å bygge boliger tar tid. Den rette linjen vil tilsvare antall boliger tilgjengelig. Etterspørsel derimot er vanskeligere å finne, men mekanismen vil være lik uansett hvordan Etterspørselskurven ser ut. Hvis antall boliger tilgjengelig, eller mengden person som leter etter en bolig øker, vil prisen gå opp. 

På lang sikt vil tilbudskurven også være diagonal. Dette kan bli forklart ved at for en hvis pris vil boligbyggere vurdere at det er verdt å bygge nye boliger. Hvis markedsprisen er lav vil dette føre til at færre boliger bli bygget. I motsatt retning vis prisen er høy vil det bli flere boliger bygget. Et eksempel på dette ser vi under 

```python
plt.plot(demand(p_num),color = "green")
plt.axvline(x = 7, color = "orange")
plt.axvline(x = 10, color = "red")

plt.xlim(0,20)


plt.plot ([0,7], [9.25, 9.25], linestyle='dashed', color='black')
plt.plot ([0,10], [8.95, 8.95], linestyle='dashed', color='black')

plt.annotate("Tilbud",xy=(5, 8.5), annotation_clip=False,color = "orange")
plt.annotate("Ny tilbud",xy=(10.5, 8.5), annotation_clip=False,color = "red")
plt.annotate("Etterspørsel",xy=(3, 9.7), annotation_clip=False,color = "green")
plt.annotate("Pris",xy=(.2, 9.3), annotation_clip=False)
plt.annotate("Ny pris",xy=(.2, 9), annotation_clip=False)

plt.ylabel('Pris', loc ='top', rotation = 0)
plt.xlabel('Kvantum', loc = 'right')

plt.title('Markedslikevekt figur ved statisk tilbud')
```

På lang sikt vil tilbudskurven også være diagonal. Dette kan bli forklart ved at for en hvis pris vil boligbyggere vurdere at det er verdt å bygge nye boliger. Hvis markedsprisen er lav vil det være mindre verdt å bygge boliger dermed vil færre boliger bli bygget. I motsatt retning vis prisen er høy vil det føre til større verdier å dermed flere boliger bygget. denne mekanismen bestemmer hva likevektsprisen i markedet er. Et eksempel på dette ser vi under 

```python
plt.plot(demand(p_num),color = "green")
plt.plot(supply(p_num),color = "orange")
plt.plot(new_supply(p_num),color = "red")



plt.xlim(0,20)
plt.ylim(8,10)


plt.plot ([0,7.5], [9.2, 9.2], linestyle='dashed', color='black')
plt.plot ([0,9.5], [9, 9], linestyle='dashed', color='black')

plt.annotate("Tilbud",xy=(3.5, 8.5), annotation_clip=False,color = "orange")
plt.annotate("Ny tilbud",xy=(7, 8.5), annotation_clip=False,color = "red")
plt.annotate("Etterspørsel",xy=(3, 9.7), annotation_clip=False,color = "green")
plt.annotate("Pris",xy=(.2, 9.25), annotation_clip=False)
plt.annotate("Ny pris",xy=(.2, 9.05), annotation_clip=False)

plt.ylabel('Pris', loc ='top', rotation = 0)
plt.xlabel('Kvantum', loc = 'right')

plt.title('Markedslikevekts figur ved dynamisk tilbud')
```

### Mulige tiltak 
I denne delen skal jeg legge frem forslag til tiltak som kan løse problemstillingene i boligmarkedet som ble identifisert tidligere. Det vil kunne være andre mulige tiltak.

#### overordnet mål med tiltakene
For å senke prisen i følge tilbuds å etterspørselsmodellen er hovedkonseptet relativt simpelt. Det innebærer å enten senke mengde etterspørsel, dette ville være negativt hvis det er ønskelig å øke befolkningen i området. Den andre muligheten vil være å øke mengden boliger. Dette kan bli gjort på flere måter, noen eksempler på dette kan være.

#### Forslag til tiltak
-Regulere flere tomter til boligtomter, dette vil føre til en redusert pris og flere muligheter for bolig bygging. Noe som på lang tid vil øke mengden boliger å dermed senke boligprisen. Dette kan ikke være ønskelig siden det kan stride med den overordredete planen for byen. I tillegg kan det gå ut over grønne områder og øke befolkningstetthet, dette kan gå utover befolkningens glede og mentale helse.

-Senke eventuelle skatter på bygging av boliger, dette vil gjøre å bygge boliger mere attraktivt å dermed øke antall boliger. Det kan samtidig senke mengden skatt staten tar inn. Som dermed kan føre til at staten ikke har penger til eventuelle goder og nødvendigheter staten gir til befolkningen, for eksempel helsehjelp og offentlig transport.

-Senke regulerings krav til byggede boliger. Det kan øke hastigheten det tar å bygge boliger, samt redusere kostnaden som vil gjøre det mer attraktivt. Potensielle negative effekter kan være at kvaliteten på boliger går ned, i tillegg til tryggheten til boliger. I tillegg kan det gjøre at det er færre boliger som fyller kravene til for eksempel rullestol brukere. 



### Konklusjon
Priser på boligmarkedet har økt drastisk fra 2019 til 2023, i samme tid har antall boliger solgt årlig gått ned. Dette har konsekvenser for arbeidsmarkedet og befolkningen i helhet, noe som kan føre til en negativ effekt på samfunnet som helhet. Det er flere mulige tiltak som kan hjelpe, de er fordeler å bakdeler med alle tiltakene. Tiltakene er alle basert rundt ideen om å gjøre det lettest mulig for utbyggere å bygge nye boliger, dette kan bli gjort med direkte hjelp eller å fjerne hindringer.
